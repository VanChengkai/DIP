# 霍夫变换(Hough Transform)

```python
import numpy as np 
from scipy import ndimage 
from matplotlib import pyplot as plt 
```

```python
# 加载canny边缘检测的机场图
f = plt.imread("./airport_edge.png")
f = f.copy()
f = f[:,:,1]
plt.imshow(f,"gray")
```

```python
# 霍夫变换
## 投票
h,w = f.shape
max_rho = max(h,w)
delta_theta = 1
steps = int(180  / delta_theta) 
vote = np.zeros((steps , max_rho))
has_value_y , has_value_x = np.where(f != 0) 
for step in range(steps): 
    for rho in range(max_rho):
        theta = step * 180 / np.pi 
        cos_theta = np.cos(theta)
        sin_theta = np.sin(theta)
        for y in has_value_y:
            for x in has_value_x:
                value = f[y][x]
                if value == 0 : 
                    continue 
                if(x*cos_theta + y * sin_theta - rho < 0.001):
                    vote[step][rho] += 1
```

```python
plt. imshow(vote , "gray")
```
