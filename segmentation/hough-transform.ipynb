# 霍夫变换(Hough Transform)

```python
import numpy as np 
from scipy import ndimage 
from matplotlib import pyplot as plt 
from skimage import feature , transform
```

```python
# 加载canny边缘检测的机场图
f = plt.imread("./../images/airport.tif")

f = feature.canny(f , sigma = 3)

plt.imshow(f,"gray")
```

```python
# 霍夫变换
## 投票
import math
h,w = f.shape
max_rho = int(math.sqrt(h ** 2 + w** 2))
delta_theta = 1
steps = int(180  / delta_theta) 
vote = np.zeros((max_rho , steps))
has_value_y , has_value_x = np.where(f != 0) 

for i in range(len(has_value_y)):  
    y = has_value_y[i]
    x = has_value_x[i]   
    for theta in range(steps):
        t = np.pi / 180 * theta 
        rho = int(x * np.cos(t) + y * np.sin(t)) - 1
        vote[rho][theta] += 1
     
```

```python
plt. imshow(vote.T , "gray")
```
