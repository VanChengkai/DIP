```python

import numpy as np 
from matplotlib import pyplot as plt
```

## Original Picture

```python

img = plt.imread("./../images/lena_gray_256.tif")
plt.imshow(img , cmap='gray')
img = img.copy()
```

## Pepper Noise

```python
def pepper(x,percent):
    h,w = x.shape
    for i in range(h):
        for j in range(w):
            if(np.random.rand() > percent):
                continue
            x[i][j] = 0 if(np.random.rand() > 0.5) else 255
    
```

```python
pepper(img , 0.01)
plt.imshow(img ,  cmap='gray')
```

```python
pepper(img , 0.1)
plt.imshow(img ,  cmap='gray')
```
